```python
import pandas as pd
```

```python
train = pd.read_csv("package/train.csv")
train["feature 0"] = pd.to_datetime(train["feature 0"])
train["day"] = train["feature 0"].dt.day
train["month"] = train["feature 0"].dt.month
train["year"] = train["feature 0"].dt.year
train
```

```python
test = pd.read_csv("package/test.csv")
test["feature 0"] = pd.to_datetime(test["feature 0"])
test["day"] = test["feature 0"].dt.day
test["month"] = test["feature 0"].dt.month
test["year"] = test["feature 0"].dt.year
test
```

```python
sub = pd.read_csv("package/submission.csv")
sub
```

```python
X_train = train[['day', 'month', 'year', 'feature 1', 'feature 2', 'feature 3',
       'feature 4', 'feature 5', 'feature 6', 'feature 7', 'feature 9']].values
y_train = train.predict.values
X_test = test[['day', 'month', 'year', 'feature 1', 'feature 2', 'feature 3',
       'feature 4', 'feature 5', 'feature 6', 'feature 7', 'feature 9']].values
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
import sklearn
```

```python
def predictor(clf, name):
    clf.fit(X_train, y_train)
    sub.predict = clf.predict(X_test).round(1)
    sub.to_csv(f"submission_{name}.csv", index=False)
```

```python
clf = SVR(kernel="linear")
predictor(clf, "linear_svm")
```

```python
clf = SVR(kernel="poly")
predictor(clf, "poly_svm")
```

```python
clf = SVR(kernel="rbf")
predictor(clf, "rbf_svm")
```

```python
clf = LinearRegression()
predictor(clf, "linreg")
```

```python
clf = Ridge()
predictor(clf, "ridge")
```

```python
clf = Lasso()
predictor(clf, "lasso")
```

```python
clf = RandomForestRegressor()
predictor(clf, "rf")
```

```python
clf = sklearn.neural_network.MLPRegressor()
predictor(clf, "mlp")
```
